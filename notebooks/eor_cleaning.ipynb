```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the csv file - got a dtype error, used low_memory = False since this notebook is just for cleaning
# and a different csv will be used
smile_on = pd.read_csv('../data/Smileon.2021.10.17.csv', header=None, low_memory = False)
```

```python
#got this from stack exhange - take a row and make it the header
new_header = smile_on.iloc[2] #grab the first row for the header
smile_on = smile_on[3:] #take the data less the header row
smile_on.columns = new_header #set the header row as the df header
smile_on.head()
```

```python
smile_on.shape
```

```python
smile_on.reset_index(inplace=True, drop=True)
```

```python
smile_on.head()
```

```python
# drop all columns that only have null values
smile_on = smile_on.dropna(axis=1,how='all')
#drop columns with information that is not useful
smile_on = smile_on.drop('OrgNum', axis=1)
smile_on = smile_on.drop('SMILE ON 60+ Treatment Plan - Action: Upload After Picture on Finish Tab', axis=1)
smile_on = smile_on.drop('SMILE ON 60+ Clinic Information - Action: Upload Before Picture on Finish Tab', axis=1)
smile_on = smile_on.drop('CountryName', axis=1)
smile_on.shape
```

```python
# regex to change strings in column names
# can also remove plus with ('SMILE ON 60\\+')
smile_on.columns = smile_on.columns.str.replace(r'SMILE ON 60\+ ', '')
smile_on.columns = smile_on.columns.str.replace('Oral Care Encounter - ', 'oec_')
smile_on.columns = smile_on.columns.str.replace(' - Quantity', '')
smile_on.columns = smile_on.columns.str.replace('Demographic Information - ', 'di_')
smile_on.columns = smile_on.columns.str.replace('Base-line Oral and Conditions Questions - ', 'bocq_')
smile_on.columns = smile_on.columns.str.replace('Screening - ', 'scrn_')
smile_on.columns = smile_on.columns.str.replace('Treatment Plan - ', 'tp_')
smile_on.columns = smile_on.columns.str.replace('Presumptive Eligibility - ', 'pe_')
smile_on.columns = smile_on.columns.str.replace('Case Management Information - ', 'cmi_')
smile_on.columns = smile_on.columns.str.replace('Clinic Information - ', 'ci_')
smile_on.columns = smile_on.columns.str.replace('Verification of Eligibility - ', 'vi_')
smile_on.columns = smile_on.columns.str.replace('Site Information - ', 'si_')
```
