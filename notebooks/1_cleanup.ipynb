```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

Read in CSV

```python
master = pd.read_csv('../data/Smileon.2021.10.17.csv', low_memory=False, skiprows=2)
master.head()
```

```python
master = master.dropna(axis=1,how='all')
master.head()
```

```python
master = master.drop(columns=['LinkedToCallReportNum', 'PhoneType', 'OrgNum'])
```

```python
master.head()
```

OK, columns have been dropped

```python
master.dtypes
```

```python
master['CallDateAndTimeStart'] =  pd.to_datetime(master['CallDateAndTimeStart'])
```

```python
master['month'] = master['CallDateAndTimeStart'].dt.month
```

```python
lines = master.plot.line(x=master['month'], y=master.count(master['month']))
```

Percentage of missing values

```python
percent_missing = master.isnull().sum() * 100 / len(master)
percent_missing.h()
```

CHANGE COLUMN NAMES TO BE EASIER TO READ

```python
master.columns = master.columns.str.replace('Demographic Information - ', 'di_')
```

```python
master
```

```python
master['SMILE ON 60+ Site Information - Enroller ID'].value_counts()
```

```python
master.columns = master.columns.str.replace(r'SMILE ON 60\+ ', '')
master.columns = master.columns.str.replace('Oral Care Encounter - ', 'oec_')
master.columns = master.columns.str.replace(' - Quantity', '')
master.columns = master.columns.str.replace('Demographic Information - ', 'di_')
master.columns = master.columns.str.replace('Base-line Oral and Conditions Questions - ', 'bocq_')
master.columns = master.columns.str.replace('Screening - ', 'scrn_')
master.columns = master.columns.str.replace('Treatment Plan - ', 'tp_')
master.columns = master.columns.str.replace('Presumptive Eligibility - ', 'pe_')
master.columns = master.columns.str.replace('Case Management Information - ', 'cmi_')
master.columns = master.columns.str.replace('Clinic Information - ', 'ci_')
master.columns = master.columns.str.replace('Verification of Eligibility - ', 'vi_')
master.columns = master.columns.str.replace('Site Information - ', 'si_')
```
